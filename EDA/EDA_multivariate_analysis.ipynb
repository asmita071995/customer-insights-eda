```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("/content/ecommerce_data_cleaned.csv")
```

```python
df.head()
```

###Price vs Quantity

```python
sns.scatterplot(data=df, x='Product_Price_USD', y='Quantity')
plt.title("Product Price vs Quantity Purchased")
plt.show()
```

###Observation

*    Higher-priced items are generally bought in smaller quantities  indicating price sensitivity.
*    Most purchases occur at lower price points, often with single quantities



###Product Category vs Total Purchase Amount

```python
sns.boxplot(data=df, x='Product Category', y='Total_Purchase_Amount_INR')
plt.xticks(rotation=45)
plt.title("Total Purchase Amount by Product Category")
plt.show()
```

###Observation

*   Categories like Electronics may show a wider price spread, with higher outliers.
*   Categories like Books or Clothing tend to have lower average purchase amounts.



### Payment Method vs Returns

```python
pd.crosstab(df['Payment Method'], df['Returns']).plot(kind='bar', stacked=True)
plt.title("Returns by Payment Method")
plt.ylabel("Number of Orders")
plt.show()
```

###Observation

*   Some payment methods (like Cash on Delivery) may have a higher return rate, possibly due to lack of commitment at purchase.




###Gender vs Product Category

```python
sns.countplot(data=df, x='Product Category', hue='Gender')
plt.xticks(rotation=45)
plt.title("Product Category Preferences by Gender")
plt.show()
```

###Observation



* Certain categories (e.g., Clothing) may be more popular among one gender, showing a trend in consumer interest.




###Age vs Total Purchase Amount

```python
plt.hexbin(df['Customer Age'], df['Total_Purchase_Amount_INR'], gridsize=30, cmap='Blues')
plt.colorbar(label='Count')
plt.xlabel('Customer Age')
plt.ylabel('Total Purchase Amount (INR)')
plt.title('Age vs Purchase Amount (Hexbin)')
plt.show()

```

###Observation

*    Younger customers may have lower purchase amounts, while middle-aged buyers tend to spend more  likely due to higher income or family purchases.




###Churn vs Total Purchase Amount

```python
sns.boxplot(data=df, x='Churn', y='Total_Purchase_Amount_INR')
plt.xticks([0, 1], ['Retained', 'Churned'])
plt.title("Purchase Amount by Churn Status")
plt.show()
```

###Observation


*   Retained customers often have higher purchase amounts, showing their value to the business.
